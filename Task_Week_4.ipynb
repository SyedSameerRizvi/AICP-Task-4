Q.5: Figure out which exact item contains a high quantity for ['Calories', 'Total Fat', 'Carbohydrates',
'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium
(% Daily Value)', 'Iron (% Daily Value)'.


```python
get_col2 = ['Item','Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber',
                    'Sugars', 'Protein', 'Vitamin A (% Daily Value)',
                    'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)',
                    'Iron (% Daily Value)']
df = pd.read_csv("menu.csv", usecols=get_col2)

mi = {}
for column in get_col2:
    mr = df[df[column] == df[column].max()]
    inm = mr['Item'].values[0]
    mi[column] = (inm, df[column].max())

for att, (item, value) in mi.items():
    print(f"{att}: {item} - {value}")
```

Q.7: Draw a horizontal bar graph for items in each category against the calories.
Here is one sample for all items in Beef & Pork, you have to do it for items in each category.

```python
get_col3 = ['Item','Category','Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber',
                    'Sugars', 'Protein', 'Vitamin A (% Daily Value)',
                    'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)',
                    'Iron (% Daily Value)']
df = pd.read_csv("menu.csv",usecols = get_col3)
cg = df.groupby('Category')

for ct, dt in cg:
    sd = dt.sort_values(by='Calories', ascending=True)
    plt.figure(figsize=(15, 7))
    sns.barplot(x='Calories', y='Item', data=sd, palette='viridis')
    plt.title(f'Calories in {ct}')
    plt.xlabel('Calories')
    plt.ylabel('Item')
    plt.show()
print("\nOBSERVATIONS\n")
print("Calories in beef and pork: ")
print(" Beef & Pork:")
print("Among the items in the Beef & Pork category, the 'Double Quarter Pounder with Cheese' has the highest calorie content.\n")
print("Beverages:")
print("In the Beverages category, the 'Minute Maid Orange juice (Large)' has the highest calorie content.\n")
print("Breakfast:")
print("Within the Breakfast category, items such as 'Big Breakfast with Hotcakes (Regular Biscuit)' have high calorie counts.\n")
print("Chicken & Fish:")
print("The 'Chicken McNuggets (40 piece)' have the highest calorie content in the Chicken & Fish category.\n")
print("Coffee & Tea:")
print("The 'Frappe Chocolate Chip (large)' have the highest calorie content in the Coffee & Tea category.\n")
print("Desserts:")
print("The 'Hot Caramel Sundae' has the highest calorie content in the Desserts category.\n")
print("Salads:")
print("The 'Premium Southwest Salad with Crispy Chicken' has the highest calorie content in the Salads category.\n")
print("Snackes and Sides:")
print("The 'Large French Fries' have the highest calorie content in the snacks and side category")
```
